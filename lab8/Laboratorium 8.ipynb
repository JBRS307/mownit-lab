## $f(x)=x^3-5x$

- $f(x)=x^3-5x$
- $f'(x)=3x^2-5$
- $f''(x)=6x$

Pierwiastkiem najbliższym w otoczeniu $1$ jest $0$. Metoda Newtona zawodzi, ponieważ, jeśli $x_0=1$ to $x_1=-1$, a $x_2=1$ przez co schemat iteracyjny nie jest zbieżny. Wystarczy zmienić wartość $x_0=\frac{1}{3}$

```python
def f(x):
    return x**3 - 5*x

def fprime(x):
    return 3*x*x - 5

def tangent(x0, x):
    return fprime(x0) * (x - x0) + f(x0)
```

```python
points = np.linspace(start=-3, stop=3, num=600)
y_points = f(points)
y_tangent = tangent(1, points)
```

```python
plt.plot(points, y_points, label="f", color="red")
plt.plot(points, y_tangent, label="tangent", color="blue")
plt.plot(points, [0 for _ in range(600)], color="black")
plt.legend()
plt.show()
```

**Wyk. 1** - $f(x)$ ze styczną w punkcie $x_0$

```python
print(f"Root of function f in neighbourhood of 1/3: {spo.newton(f, 1/3, fprime=fprime)}")
```

$f(x)=x^2-3x+2=0$ ma równoważny schamat iteracyjny z 

 $g_1(x)=\frac{x^2+2}{3}$
 
 $g_2(x)=\sqrt{3x-2}$
 
 $g_3(x)=3-\frac{2}{x}$
 
 $g_4(x)=\frac{x^2-2}{2x-3}$

## podpunkt a
$g_1(x)$  
$x^*=g_1(x^*)$ i $|g_1'(x^*)|=|\frac{2x^*}{3}|=\frac43>1$, zatem schemat iteracyjny jest rozbieżny  
$g_2(x)$  
$x^*=g_2(x^*)$ i $|g_2'(x^*)|=|\frac{3}{2\sqrt{3x^*-2}}|=\frac34<1$, zatem schemat iteracyjny jest zbieżny. Rząd zbiezności liniowy ze stałą $C=\frac34$  
$g_3(x)$  
$x^*=g_3(x^*)$ i $|g_3'(x^*)|=|\frac2{{x^*}^2}|=\frac24=\frac12<1$, zatem schemat iteracyjny jest zbieżny. Rząd zbiezności liniowy ze stałą $C=\frac12$  
$g_4(x)$  
$x^*=g_4(x^*)$ i $|g_4'(x^*)|=|\frac{2 ({x^*}^2 - 3x^* +2)}{(2x^* - 3)^2}|=0 < 1$, zatem schemat iteracyjny jest zbieżny. Rząd zbiezności kwadratowy

## podpunkt b

```python
from numpy import log as ln
def f(x):
    return x**2 - 3*x + 2
def g1(x):
    return (x**2 + 2) / 3
def g2(x):
    return np.sqrt(3*x - 2)
def g3(x):
    return 3-(2/x)
def g4(x):
    return (x**2 - 2)/(2*x-3)
```

```python
ITER = 10
fun = [g1, g2, g3, g4]
x0 =  [2.5,2.5,2.5,2.5]
x_ks = [ [0]*ITER for _ in range(len(fun))]

for g in range(4):
    r = 0
    x_k = x0[g]
    try:
        for i in range(ITER):
            x_k = fun[g](x_k)
            x_ks[g][i] = x_k
    except:
        print(g+1, "cannot be calculated")
    print(g+1,x_k)

#abs error
def e1(k):return abs(x_ks[0][k] - 2)
def e2(k):return abs(x_ks[1][k] - 2)
def e3(k):return abs(x_ks[2][k] - 2)
def e4(k):return abs(x_ks[3][k] - 2)
#relative error
def re1(k):return e1(k)/2
def re2(k):return e2(k)/2
def re3(k):return e3(k)/2
def re4(k):return e4(k)/2

k = ITER - 2
r_1 = ln(e1(k)/e1(k+1))/ln(e1(k-1)/e1(k))# g1 diverges
r_2 = ln(e2(k)/e2(k+1))/ln(e2(k-1)/e2(k))
r_3 = ln(e3(k)/e3(k+1))/ln(e3(k-1)/e3(k))
k = 3
r_4 = ln(e4(k)/e4(k+1))/ln(e4(k-1)/e4(k))
print("R for g1:",r_1,"R for g2:",r_2,"R for g3:",r_3,"R for g4:",r_4)
```

Zatem dane eksperymentalne potwierdzily rozważania teoretyczne i g1 jest rozbiezny, a g2 oraz g3 maja liniowy rzad zbiezności. Natomiast g4 cechuje się zbieżnościa kwadratową

# Zadanie 4

$$
\left\{
\begin{aligned}
x_1^2+x_2^2=1\\
x_1^2-x_2=0
\end{aligned}
\right.
$$

$$
\left\{
\begin{aligned}
x_1^2+x_2^2-1=0\\
x_1^2-x_2-0=0
\end{aligned}
\right.
$$

$$
\textbf{f(x)}=
\begin{bmatrix}
x_1^2 & +x_2^2 & -2\\
x_1^2 & -x_2 & + 0
\end{bmatrix}=\textbf{0}
$$

$$
\textbf{J}=\begin{bmatrix}
2x_1 & 2x_2\\
2x_1 & -1
\end{bmatrix}
$$

$$
J(x_k)\cdot S_k=-f(x_k)\\
x_{k+1} =x_k+s_k
$$

```python
x_0 = np.array([5,5]).T
x_k = x_0
x_ks = [0] * 10
def J(x): return np.array([[2*x[0], 2*x[1]],[2*x[0], -1]])
def f(x): return np.array([x[0]**2+x[1]**2 -1,x[0]**2 - x[1]])
for i in range(10):
    s_k = np.linalg.solve(J(x_k), -f(x_k))
    x_k = x_k + s_k
    x_ks[i] = x_k
print("Result:", x_k)
```

```python
exac_x = np.array([np.sqrt((np.sqrt(5)/2)-(1/2)), (np.sqrt(5)/2)-(1/2)])
print("Actual result:", exac_x)
```

```python
print("Relative Error",np.array([100 * (abs(x_k[0] - exac_x[0])/exac_x[0]),  100 * (abs(x_k[1] - exac_x[1])/exac_x[1])]), "[%]")
```

```python
x_k_rel_error = [[0]*2 for _ in range(10)]
for i in range(10):
    x_k_rel_error[i][0] = (abs(x_ks[i][0] - exac_x[0]) / exac_x[0]) * 100
    x_k_rel_error[i][1] = (abs(x_ks[i][1] - exac_x[1]) / exac_x[1]) * 100
```

```python
plt.yscale("log")
plt.xlabel("Number of Iterations")
plt.ylabel("Relative error [%]")
plt.plot(np.arange(10), x_k_rel_error, label="relative error", color="blue")
plt.show()
```
