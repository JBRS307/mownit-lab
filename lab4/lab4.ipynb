```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f1(x):return(1/(1+25*x**2))
def f2(x):return np.exp(np.cos(x))
```

```python
range_1 = np.linspace(-1,1,500)
range_2 = np.linspace(0,2*np.pi,500)
```

```python
plt.title("f1(x)")
plt.xlabel("X") 
plt.ylabel("Y")
plt.plot(range_1, f1(range_1), color ="green")
plt.show()
```

```python
plt.xlabel("X")
plt.title("f2(x)")
plt.ylabel("Y")
plt.plot(range_2, f2(range_2), color ="blue")
plt.show()
```
